```python
res = sns.barplot(x='cyl',y='mpg',data=mtcars,color='green')
plt.show()
```

```python
res = sns.barplot(x='am',y='mpg',data=mtcars)
plt.show()
```

```python
res = sns.barplot(x='gear',y='hp',data=mtcars)
for bars in res.containers:
    res.bar_label(bars)
plt.show()
```

```python
ax=sns.countplot(x='cyl',data = mtcars,hue = 'cyl')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(x='carb',data = mtcars,hue = 'am')
```

```python
ax = sns.countplot(y='gear',data = mtcars , hue='gear',legend=False)
sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(style="ticks")
sns.pairplot(mtcars[["mpg", "wt", "hp"]], diag_kind="kde")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x="am", data=mtcars)
plt.title("Transmission Type (AM)")
plt.xlabel("Transmission Type (0 = Automatic, 1 = Manual)")
plt.ylabel("Count")
plt.show()
```

```python
g = sns.FacetGrid(mtcars, col="cyl", col_order=[4, 6, 8])
g.map(plt.scatter, "wt", "mpg")
plt.show()
```
